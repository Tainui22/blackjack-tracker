# Blackjack Team Performance Dashboard
Visualize session results, splits, streaks, and best strategies.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('session_data.csv')
df['SB_Team'] = df['Total_Profit'] * 0.7
df['Casino'] = df['Total_Profit'] * 0.3

# Plot cumulative team/casino profits
plt.figure(figsize=(8,5))
plt.plot(df['Session_Date'], df['SB_Team'].cumsum(), label='SB-Team')
plt.plot(df['Session_Date'], df['Casino'].cumsum(), label='Casino Opponents')
plt.legend()
plt.title('Cumulative Profits')
plt.xlabel('Session Date')
plt.ylabel('Profit ($)')
plt.show()

# Best strategy analysis
best_strategy = df.groupby('Strategy_Used')['Total_Profit'].mean().idxmax()
print(f"Best strategy so far: {best_strategy}")
```
